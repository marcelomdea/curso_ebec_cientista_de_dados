#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# %matplotlib inline
%matplotlib notebook

dados = pd.read_csv('sinasc_RO_2019.csv')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(3, 2)
plt.subplots_adjust(hspace=0.3)

axes[2, 0].tick_params(axis="x", rotation = 15)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=dados)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=dados)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=dados)
sns.countplot(ax = axes[1, 1], x='SEXO', data=dados)
sns.countplot(ax = axes[2, 0], x='GESTACAO', data=dados, )
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
fig, axes = plt.subplots()

sns.barplot(ax = axes, x='CONSULTAS', y='APGAR5', data=dados)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)
plt.subplots_adjust(hspace=0.4)

axes[0, 1].tick_params(axis="x", rotation = 15)

sns.countplot(ax = axes[0, 0], x='SEXO', data=dados)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=dados)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=dados)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=dados)
```
